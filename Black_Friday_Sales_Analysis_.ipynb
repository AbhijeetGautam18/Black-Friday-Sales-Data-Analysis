```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
```

```python
# Total sales during the Black Friday Sale
a = df['Purchase'].sum()
print(f'Total Sales during Black Friday Sale was {a}')
```

```python
# Sales distribution across different product categories.

df1 = df.groupby(['Product_Category_1'],as_index=False).agg(Total_Sales = ('Purchase','sum'))
df1
```

```python
plt.pie(df1['Total_Sales'],labels = df1['Product_Category_1'])
```

```python
df.head()
```

```python
# Understand the demographics of your customers, including gender, age, occupation, city category, marital status, and stay in the current city.

age_purchase = df.groupby(['Age']).agg(Purchase = ('Purchase','sum'))
age_purchase
```

```python
gender_purchase = df.groupby(['Gender']).agg(Purchase = ('Purchase','sum'))
gender_purchase
```

```python
city_purchase = df.groupby(['City_Category']).agg(Purchase = ('Purchase','sum'))
city_purchase
```

```python
marital_purchase = df.groupby(['Marital_Status']).agg(Purchase = ('Purchase','sum'))
marital_purchase
```

```python
city_stay_purchase = df.groupby(['City_Category','Stay_In_Current_City_Years','Age','Gender','Marital_Status']).agg(Purchase = ('Purchase','sum'))
city_stay_purchase[city_stay_purchase['Purchase'] == city_stay_purchase['Purchase'].max()]
```

```python
# Product Analysis:

# Identify the top-selling products.
# Analyze the sales distribution across different product categories


prod_sold = df.groupby(['Product_Category_1']).agg(Prod_sold = ('Product_Category_1','count'), Purchase = ('Purchase','sum'))
prod_sold
```

```python
prod_sold_seg = df.groupby(['Product_Category_1','City_Category','Stay_In_Current_City_Years','Age','Gender','Marital_Status']).agg(Prod_sold = ('Product_Category_1','count'), Purchase = ('Purchase','sum'))
prod_sold_seg
```

```python
prod_sold_seg[prod_sold_seg['Prod_sold'] == prod_sold_seg['Prod_sold'].max()]
```

```python
prod_sold_seg[prod_sold_seg['Purchase'] == prod_sold_seg['Purchase'].min()]
```

```python
# Purchase Behavior:

# Analyze the average purchase amount per customer.

avg_purchase = df['Purchase'].sum()/df['User_ID'].nunique()
avg_purchase
```
